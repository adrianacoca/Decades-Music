```python
import pandas as pd
import numpy as np
```

```python
fif = pd.read_csv('src/1950.csv')
six = pd.read_csv('src/1960.csv')
sev = pd.read_csv('src/1970.csv')
eig = pd.read_csv('src/1980.csv')
nin = pd.read_csv('src/1990.csv')
zer = pd.read_csv('src/2000.csv')
ten = pd.read_csv('src/2010.csv')
fif.head(10)
```

```python
fif["year"]=1950
six["year"]=1960
sev["year"]=1970
eig["year"]=1980
nin["year"]=1990
zer["year"]=2000
ten["year"]=2010


```

```python
alldecades = pd.concat([fif, six, sev, eig, nin, zer, ten])
```

```python
alldecades = alldecades.drop(["bpm", "nrgy", "dB", "live", "acous", "spch"], axis=1)
```

```python
alldecades.shape
```

```python
alldecades.reset_index(drop=True)
```

```python
alldecades.artist.value_counts().head()
```

```python
alldecades = alldecades.rename(columns={"Number": "Number", "title": "Title", "artist": "Artist", "top genre": "Genre", "year": "Year", "dnce": "Dance", "pop": "Popularity", "dur": "Duration", "val": "Positiveness"})
```

```python
alldecades.to_csv('output/music_decades.csv')
```
