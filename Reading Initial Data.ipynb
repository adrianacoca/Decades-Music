```python
def getFromGithub(endpoint, queryParams={}):
    url = f"https://accounts.spotify.com/api/token{endpoint}"
    apiKey = base64.b64encode(os.getenv("client_secret"))
    clientID = base64.b64encode(os.getenv("client_id"))
    if not apiKey:
        raise ValueError("Please setup the GITHUB_APIKEY env variable")
    headers = {
        "Authorization": f"token {apiKey}"
    }
    
    res = requests.get(url, headers=headers, params=queryParams)
    print(res.url)
    return res.json()
```

```python
#try2
def getFromSpotify(queryParams={}):
    url = f"https://accounts.spotify.com/authorize"

    apiKey = base64.b64encode(os.getenv("client_secret"))
    clientID = base64.b64encode(os.getenv("client_id"))
    
    if not apiKey:
        raise ValueError("Please setup the SPOTIFY_APIKEY env variable")
    headers = {
        "Authorization" : f"{clientID}:{apiKey}"
    }
    
    res = requests.get(url, headers=headers, params=queryParams)
    print(res.url)
    print(apiKey, clientID)
    return res.json()
```

```python
#try3 - el bueno
def get_spotify_oauth_token(token_url):
    client_secret = os.getenv("client_secret")
    client_id = os.getenv("client_id")
    to_encode = f"{client_id}:{client_secret}"
    encoded_auth = base64.b64encode(to_encode.encode())
    headers = {"Authorization": "Basic %s" % encoded_auth}
    data = {'grant_type': "client_credentials"}
    response = requests.post(url=token_url, headers=headers, data=data)

    if response.status_code == requests.codes.ok:
        return response.json()['access_token']
    else: 
        response_error = json.loads(response.text).get('error')
        msg = "Error getting OAuth token: %s" % response_error
        raise Exception(msg)
        
"authorization_code"
/authorize endpoint
redirect_uri
```
